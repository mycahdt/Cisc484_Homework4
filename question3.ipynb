```python
import sklearn as sk
from sklearn import svm
import pandas as pd
from sklearn.svm import SVC
```

```python
file= open("q4_training.txt")
data_text= file.read()
lines= data_text.split("\n")
data_training_x= []
data_training_y=[]
for line in lines:
    if len(line)>0:
        arr=line.split()
        data_training_y.append(float(arr[0]))
        data_training_x.append([float(arr[1]), float(arr[2])])
print(data_training_x)
print(data_training_y)
```

```python
file= open("q4_test.txt")
data_text2= file.read()
lines2= data_text2.split("\n")
data_test_x= []
data_test_y=[]
for line in lines2:
    if len(line)>0:
        arr=line.split()
        data_test_y.append(float(arr[0]))
        data_test_x.append([float(arr[1]), float(arr[2])])
print(data_test_x)
print(data_test_y)
```

```python
'''
3a.)
With linear and quadratic kernels and default choice for C parameter,
save the output for test data and plot using plot.py.
Report the number of errors or accuracy on the test set for each case.
'''
m1= svm.SVC(kernel='linear')
m1.fit(data_training_x,data_training_y)
predict_m1= m1.predict(data_test_x)
print(predict_m1)

m2= svm.SVC(kernel='poly', degree=2)
m2.fit(data_training_x,data_training_y)
predict_m2= m2.predict(data_test_x)
print(predict_m2)

```

```python
from sklearn.metrics import accuracy_score
print("Accuracy for m1 (kernel=linear): ", accuracy_score(data_test_y, predict_m1))
print("Accuracy for m2 (kernel=poly): ", accuracy_score(data_test_y, predict_m2))

print("Error for m1 (kernel=linear): ", 1-accuracy_score(data_test_y, predict_m1))
print("Error for m2 (kernel=poly): ", 1-accuracy_score(data_test_y, predict_m2))
```

```python
import sys
import matplotlib.pyplot as pl

try:
	DATASET = sys.argv[1]
	PREDICT = sys.argv[2]
except:
	print("Error! Use the command: python plot.py [Test_set] [prediction_file]")
################################################################################

# Read dataset and return 3 lists corresponding to classification, x and y
def opendataset(dataset):

    fid = open(dataset,'r')

    c = []
    x = []
    y = []

    for line in fid:
        c.append(float(line.split()[0]))
        x.append(float(line.split()[1]))
        y.append(float(line.split()[2]))

    fid.close()

    return c,x,y

# Read predict output
def openpredict(predictfile):

    fid = open(predictfile,'r')

    predict = []

    for line in fid:
        predict.append(float(line))

    return predict

# Show plot
def showplot(c,x,y,predict):

    x_tp, x_tn, y_tp, y_tn = [],[],[],[]
    x_fp, x_fn, y_fp, y_fn = [],[],[],[]

    # Confusion matrix
    for i in range(len(c)):
        if c[i] > 0 and predict[i] > 0:
            x_tp.append(x[i])
            y_tp.append(y[i])
        elif c[i] < 0 and predict[i] < 0:
            x_tn.append(x[i])
            y_tn.append(y[i])
        elif c[i] < 0 and predict[i] > 0:
            x_fp.append(x[i])
            y_fp.append(y[i])
        elif c[i] > 0 and predict[i] < 0:
            x_fn.append(x[i])
            y_fn.append(y[i])
        else:
            print('Error! Check your predict file!')

    p_tp, = pl.plot(x_tp,y_tp,'ro') # True positive -> Red dot
    p_tn, = pl.plot(x_tn,y_tn,'bo') # True negative -> Blue dot
    p_fp, = pl.plot(x_fp,y_fp,'g^') # False positive -> Green triangle
    p_fn, = pl.plot(x_fn,y_fn,'ms') # False negative -> magenta square

    pl.legend([p_tp, p_tn, p_fp, p_fn],
              ['True Positive', 'True Negative','False Positive','False Negative'],
              loc = 3,
              ncol = 2,
              mode = 'expand',
              numpoints=1,
              borderaxespad=0.,
              bbox_to_anchor=(0., 1.01, 1., .102))


    pl.show()

    return None

def main():
    DATASET = "q4_test.txt"
    PREDICT = "/content/drive/MyDrive/3_Cisc484_MachineLearning/Homework4/question3a_prediction_m1.txt"
    c,x,y = opendataset(DATASET)
    p = openpredict(PREDICT)
    showplot(c,x,y,p)


    DATASET = "q4_test.txt"
    PREDICT = "/content/drive/MyDrive/3_Cisc484_MachineLearning/Homework4/question3a_prediction_m2.txt"
    c,x,y = opendataset(DATASET)
    p = openpredict(PREDICT)
    showplot(c,x,y,p)


main()
```

```python
max_acc= 0
best_c=0
for c in range(1,10000):
    m3= svm.SVC(C= c,kernel='rbf')
    m3.fit(data_training_x,data_training_y)
    predict_m3= m3.predict(data_test_x)
    #print(predict_m3)
    acc= accuracy_score(data_test_y, predict_m3)
    if acc> max_acc:
        max_acc= acc
        best_c= c
print("Best C value: "+str(best_c))
print("\nAccuracy: "+str(max_acc))
```

```python
m3= svm.SVC(C= best_c,kernel='rbf')
m3.fit(data_training_x,data_training_y)
predict_m3= m3.predict(data_test_x)
#print(predict_m3)
acc= accuracy_score(data_test_y, predict_m3)
print("\nAccuracy: "+str(max_acc))
```
